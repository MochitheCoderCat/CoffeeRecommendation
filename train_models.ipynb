```python
# Define a function to standardize prices to dollars per ounce
def convert_to_dollar_per_ounce(price):
    try:
        # Define approximate conversion rates to USD
        conversion_rates = {
            "NT $": 0.032,   # New Taiwan Dollar
            "NTD $": 0.032,
            "NTD": 0.032,
            "NT$": 0.032,
            "NT": 0.032,
            "Nt $": 0.032,
            "Nt": 0.032,
            "CAD $": 0.75,   # Canadian Dollar
            "AED ": 0.27,    # United Arab Emirates Dirham
            "AUD $": 0.65,   # Australian Dollar
            "pesos": 0.057,  # Mexican Peso
            "HK $": 0.13,    # Hong Kong Dollar
            "HKD $": 0.13,   # Hong Kong Dollar
            "IDR ": 0.000065, # Indonesian Rupiah
            "KRW ": 0.00076, # South Korean Won
            "RMB": 0.1381,   # RMB
            "RM ": 0.21,     # Malaysian Ringgit
            "#": 1
        }
        # Iterate through conversion rates to find a match in the price string
        for currency, rate in conversion_rates.items():
            if currency in price:
                amount, unit = price.replace(currency, "").split("/")
                amount = float(amount)
                if "grams" in unit:
                    # Convert grams to ounces (1 gram = 0.035274 ounces)
                    grams = float(unit.replace(" grams", ""))
                    ounces = grams * 0.035274
                elif "ounces" in unit:
                    ounces = float(unit.replace(" ounces", ""))
                elif "pounds" in unit:
                    pounds = float(unit.replace(" pounds", ""))
                    ounces = pounds * 16
                else:
                    return None  # If unit is not recognized

                # Convert to USD per ounce
                usd = amount * rate
                return usd / ounces

        # Handle direct USD prices as before
        # #
        if "$" in price:
            amount, unit = price.replace("$", "").split("/")
            amount = float(amount)
            if "ounces" in unit:
                ounces = float(unit.replace(" ounces", ""))
                return amount / ounces
            elif "grams" in unit:
                grams = float(unit.replace(" grams", ""))
                ounces = grams * 0.035274
                return amount / ounces
            elif "pounds" in unit:
                pounds = float(unit.replace(" pounds", ""))
                ounces = pounds * 16
                return amount / ounces

    except Exception:
        return None  # Return None for invalid formats



# Apply the conversion to the est_price column
df["price_per_ounce"] = df["est_price"].apply(convert_to_dollar_per_ounce)
```

```python
# Using KNN model
```

Calculation the distance between one coffee bean and all other coffee beans, return the top10 nearest neighbors.
*   For a single coffee bean: Recommend the highest-rated coffee bean among the top10 nearest neighbors.
*   For two coffee beans:
If there is an overlap between the 2 sets of top10 nearest neighbors for each inputted coffee beans, recommend the highest-rated coffee bean in the overlap.
If no overlap, combine the two sets of nearest neighbors and recommend the highest-rated coffee bean among all 20 neighbors.

```python

from sklearn.neighbors import NearestNeighbors

# Combine numerical and encoded categorical features
rating_column = 'rating'
name_column = 'name'

# Drop rows where 'rating' and 'name' is NaN
ratings = df[rating_column].reset_index(drop=True)
names = df[name_column].reset_index(drop=True)

# Fit KNN model
knn = NearestNeighbors(n_neighbors=10, metric='euclidean')
knn.fit(X)

# Find the 10 nearest coffee beans for each coffee bean
distances, indices = knn.kneighbors(X)

# Store the nearest neighbors and their ratings in a dictionary
nearest_neighbors = {}
for idx, neighbors in enumerate(indices):
    nearest_neighbors[names.iloc[idx]] = {
        "neighbors": [names.iloc[i] for i in neighbors if names.iloc[i] != names.iloc[idx]],
        "ratings": [ratings.iloc[i] for i in neighbors if names.iloc[i] != names.iloc[idx]],
    }
```

```python
from joblib import dump

# Save the trained KNN model
dump(nearest_neighbors, 'model/knn_model.joblib')
print("KNN model has been saved as 'model/knn_model.joblib'.")
```

------------ Code for application ----------------

```python
# Load the pre-exported dictionary
from joblib import load

# Load the KNN model
nearest_neighbors = load('./model/knn_model.joblib')

def recommend(user_input_names):
    if len(user_input_names) == 1:
        # Single input: Recommend the highest-rated neighbor
        coffee_name = user_input_names[0]
        neighbors = nearest_neighbors[coffee_name]["neighbors"]
        ratings = nearest_neighbors[coffee_name]["ratings"]
        highest_rated_idx = ratings.index(max(ratings))
        return neighbors[highest_rated_idx]

    elif len(user_input_names) == 2:
        # Two inputs: Check for overlap
        coffee_1, coffee_2 = user_input_names
        neighbors_1 = set(nearest_neighbors[coffee_1]["neighbors"])
        neighbors_2 = set(nearest_neighbors[coffee_2]["neighbors"])

        overlap = neighbors_1.intersection(neighbors_2)
        if overlap:
            # Recommend the highest-rated coffee in the overlap
            overlap_ratings = [(name, nearest_neighbors[name]["ratings"][0]) for name in overlap]
            return max(overlap_ratings, key=lambda x: x[1])[0]
        else:
            # Recommend the highest-rated coffee among all 20 neighbors
            combined_neighbors = list(neighbors_1.union(neighbors_2))
            combined_ratings = [(name, nearest_neighbors[name]["ratings"][0]) for name in combined_neighbors]
            return max(combined_ratings, key=lambda x: x[1])[0]

# Example usage
user_input_1 = ["Kenya Nyeri AA Ichuga"]
user_input_2 = ["Kenya Nyeri AA Ichuga", "Ethiopia Yirgacheffe"]

print("Recommended coffee bean for single input:", recommend(user_input_1))
print("Recommended coffee bean for two inputs:", recommend(user_input_2))
```

```python
# Using a different model -- K-means
```

Cluster the coffee beans based on the specified features. Each coffee bean will be assigned to a cluster.
*   For a single coffee bean: Recommend the highest-rated coffee bean in the same cluster.
*   For two coffee beans: If the two coffee beans are in the same cluster, recommend the highest-rated coffee bean in that cluster.
If they are in different clusters, combine the two clusters and recommend the highest-rated coffee bean.




```python
from sklearn.cluster import KMeans

# Combine numerical and encoded categorical features
X = pd.concat([df[numerical_features].reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)
rating_column = 'rating'
name_column = 'name'

# Drop rows where 'rating' and 'name' is NaN
ratings = df[rating_column].reset_index(drop=True)
names = df[name_column].reset_index(drop=True)

# Train the K-means model
n_clusters = 100
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# Create a dictionary where keys are coffee bean names
clusters_dict = {}
for index, row in df.iterrows():
    # Get the cluster ID and all beans in that cluster
    cluster_id = row['cluster']
    cluster_beans = df[df['cluster'] == cluster_id]
    cluster_beans.drop(cluster_beans[cluster_beans.name == row[name_column]].index, inplace=True)
    name_list = cluster_beans[name_column].tolist()
    rating_list = cluster_beans[rating_column].tolist()

    # Add the coffee bean and its cluster info to the dictionary
    clusters_dict[row[name_column]] = {
        "cluster": int(cluster_id),
        "neighbors": name_list,
        "ratings": rating_list
    }
```

```python
from joblib import dump
# Save the trained KMeans model
dump(clusters_dict, 'model/kmeans_model.joblib')
print("KMeans model has been saved as 'kmeans_model.joblib'.")
```

```python
# Recommendation logic
def recommend_kmeans(user_input_names):
    input_clusters = set()
    relevant_beans = []

    for name in user_input_names:
        # Add the cluster ID to input_clusters
        input_clusters.add(clusters_dict[name]["cluster"])

        # Add all neighbors to relevant beans
        neighbors = clusters_dict[name]["neighbors"]
        ratings = clusters_dict[name]["ratings"]
        relevant_beans.extend(zip(neighbors, ratings))

    # Filter unique beans if input clusters are different
    if len(input_clusters) > 1:
        # Deduplicate neighbors
        relevant_beans = list({bean[0]: bean for bean in relevant_beans}.values())

    # Recommend the highest-rated coffee bean
    recommended_bean = max(relevant_beans, key=lambda x: x[1])
    return recommended_bean[0]

# Example usage
user_input_1 = ["Kenya Nyeri AA Ichuga"]  # Single coffee bean name input
user_input_2 = ["Kenya Nyeri AA Ichuga", "Ethiopia Yirgacheffe"]  # Two coffee bean names input

print("Recommended coffee bean for single input:", recommend_kmeans(user_input_1))
print("Recommended coffee bean for two inputs:", recommend_kmeans(user_input_2))
```
